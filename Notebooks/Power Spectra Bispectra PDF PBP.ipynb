# Point by point

## Note it is not possible to make one of these for the hod benchmark, they consider different locations possible for galaxies!

```python
x = np.log10(test_cube.flatten())
y = np.log10(pred_cube.flatten())
```

```python
h = sns.jointplot(x, y,kind='kde')

# JointGrid has a convenience function
h.set_axis_labels('x', 'y', fontsize=16)

# or set labels via the axes objects
h.ax_joint.set_xlabel('Target Mass (log $M_{sun}$)', fontweight='bold',fontsize = 15)

h.ax_joint.set_ylabel('Predicted Mass (log $M_{sun}$)', fontweight='bold',fontsize = 15)

# also possible to manipulate the histogram plots this way, e.g.
h.ax_marg_y.grid('on') # with ugly consequences...

# labels appear outside of plot area, so auto-adjust
plt.tight_layout()
```

From this, you can vaguely see that the slope is less than one, corresponding to the consistent underprediction of stellar mass, which makes sense when you also condier the sparsity concerns. 

Some room for improvement lies in using the residual as the target for modelling,but the issue here is that since there isn't a 1:1 correspondence between where the hod predicts galaxies, we cannot always get a residual value, and end up with a huge error as a result.
